```python
target = pd.DataFrame(data['CIBIL Score'])
```

```python
data.drop('CIBIL Score',axis=1,inplace=True)
```

```python
features=data
```

```python
del features['Leads']
del features['Vertical']
del features['Repayment']
del features['Product']
del features['Loan_ID']
del features['CustID']
```

```python
features.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
features_scaled = sc.fit_transform(features)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
features_pca = pca.fit_transform(features_scaled)
```

```python
variance_ratio = pca.explained_variance_ratio_
variance_ratio
```

```python
percentage_variance = 0
for x in range(len(variance_ratio)):
    percentage_variance = variance_ratio[x] * 100 + percentage_variance
    print('principal component :' , x  , ' and variance ratio is : ' , percentage_variance )
```

```python
pca = PCA( n_components = 4)
features_pca = pca.fit_transform(features_scaled)
```

```python
from sklearn.cluster import KMeans
```

```python
k_values = range(1,20)
inertias=[]

for k in k_values :
    # create a KMeans instance with K clusters : model
    model = KMeans(n_clusters=k)
    
    # fit model to samples
    model.fit(features_pca)
    
    # append the values to list
    inertias.append(model.inertia_)
```

```python
inertias
```

```python
plt.plot(k_values , inertias)
plt.show()
```

```python
model2 = KMeans(n_clusters= 5)
```

```python
labels = model2.fit_predict(features_pca)
```

```python
labels.shape
```

```python
target['target_label'] = labels
```

```python
target.head()
```
